```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data import
survey_data=pd.read_json('survey_final_data.json')
survey_data
```

```python
survey_data['question_id'].unique()
```

```python
profile_data = survey_data.groupby(['user_id','gender','age','loc','interests'])['loc'].size().reset_index(name='questions_answered')
profile_data
```

#### Location

```python
#location = survey_data.groupby(['loc'])['user_id'].size().reset_index(name='count')
location=profile_data['loc'].unique()
location
```

#### Gender

```python
gender=profile_data.groupby(['gender']).size().reset_index(name='count')
gender
```

```python
# set seaborn style and font scale
plt.subplots(figsize=(10,7))
sns.set_style('whitegrid')
#sns.set(font_scale=1.2)

sns.countplot(x='gender', data=profile_data, palette='husl')
#sns.set_palette(['#69F960','#9CEF7C'])

plt.title('Distribution of Gender', fontsize=18)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
# plt.legend(loc='upper left')
plt.xticks(fontsize=12, rotation=45)
# add text annotations to bars
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}',
                (p.get_x() + p.get_width() / 2, p.get_height() + 0.5),
                ha='center', va='bottom', fontsize=12)
# display the plot
plt.tight_layout()
plt.show()
```

#### Age distribution

```python
#plt.hist(profile_data['age'], bins=20)

sns.kdeplot(profile_data['age'], shade=True)
sns.rugplot(profile_data['age'], color=".2", linewidth=1)

# set the title and labels
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Density")

# display the plot
plt.show()
```

#### Prefered soda

```python
soda=[7]
prefered_soda=survey_data[survey_data['question_id'].isin(soda)]
#prefered_soda.head()
```

```python
soda_count = prefered_soda['selected_choices'].value_counts()
#soda_count
# create a new dataframe with the value counts as columns
soda_prefences = pd.DataFrame({'selected_choices': soda_count.index, 'users prefered soda': soda_count.values})
soda_prefences
```
